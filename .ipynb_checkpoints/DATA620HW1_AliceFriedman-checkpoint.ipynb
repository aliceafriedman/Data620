## Part 2: Create a graph from scratch

Using the built in functions of the ```networkx``` package, we can generate a small <a href="https://en.wikipedia.org/wiki/Krackhardt_kite_graph" target="_blank">Krackhardt Kite Network</a>, similar to the one in the example given in the assignment. 



````python
G = net.generators.small.krackhardt_kite_graph() 
````

````python
print(G)
G.nodes()
````

````python
net.draw(G)
````

We can add labels to the nodes or edges by creating a ```dict``` object.

````python
labeldict = {}
labeldict[0] = "Beth"
labeldict[1] = "Ben"
labeldict[2] = "Sally"
labeldict[3] = "Bethany"
labeldict[4] = "Benjamin"
labeldict[5] = "SallyAnn"
labeldict[6] = "Bitsy"
labeldict[7] = "Benjy"
labeldict[8] = "Sal"
labeldict[9] = "Sarah"


net.draw(G, labels=labeldict, with_labels=True, node_size=80, alpha = 0.5, font_size=15)
````

A Krackhardt Kite Network is a kind of graph in which the maximum of the three measures of how connected a node is--degree, "betweenness centrality" , and "closeness centrality"--are all different points. 

"Bethany" has the greatest degree--that is, she is the most connected of anyone in the social network. This measure is the simplest to conceptutalize, and can be easily determined by a visual inspection of the graph. But although Bethany may have the most friends, in another sense she is quite distant from Sarah. That is where the other measures come in. "Sarah" has the lowest degree--she has only one friend, "Sal". This measure, however, is independent of size or configuration of the overall network, so may not always provide the most useful information about a network. 

For every pair of people, there exits a shortest chain of hops between people to connect them. The person included in the most shortest-hop-chains between pairs of people has the highest <a href="https://en.wikipedia.org/wiki/Betweenness_centrality" target="_blank">betweenness centrality</a>, meaning they are most likely to be the common link between people who are not directly friends. In the above example, "Benjy" has the highest betweeness centrality. Sarah, again, has the lower betweenness centraltiy, as she is not the common link for *any* two friends. Because this measure includes *every* connection in the network, it may be more useful for determinig the relative importance of a node.

Finally, each person can be said to be close or far to the others based on how many hops it takes to link to each one. The sum of all of these shortest changes could be said to be a measure of the "distantness" of any person. The reciprical of this sum is a measure of how "close" a person is to all the others. This measure, called, <a href="https://en.wikipedia.org/wiki/closeness_centrality" target="_blank">closeness centrality</a>, again, includes information about every connection in the network, but more heavily weights nodes withe a higher degree than the betweeness measure of centrality. Again depending on the purpose of the analysis, this measure may provide more useful in determining the most (or least!) important nodes in the network. In the above example, Bitsy and SallyAnn have the highest closeness centrality, with Sarah once again coming in last.



## Part 2: Use ```nxviz``` to present alternate visualizations

```nxviz``` is a package built on top of ```networkx``` that allows for quick data visualizations of ordered graph types, including matrices, arc plots, and a circular arc plot called a circos plot. Just for fun, let's see what we can do with it.

````python
try:
    import nxviz as nx
except: 
    import sys
    !{sys.executable} -m pip install nxviz
    import nxviz as nx
````

First, we create an object ```AP``` using the built-int functions of ```nxviz```, then we can use the draw method on that object. The same process can be used to create a CircosPlot and a MatrixPlot.

````python
AP = nx.ArcPlot(G)
AP.draw()
````

````python
CP = nx.CircosPlot(G)
CP.draw()
````

````python
M = nx.MatrixPlot(G)
M.draw()
````
